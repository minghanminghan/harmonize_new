```python
'''
preprocessing goals:
1. read all data into one df
2. drop duplicates (by id/uri) and columns that don't appear in every dataset
3. write data to csv & delete old csv's
4. create knn model and dump
'''

import os
import numpy as np
import pandas as pd

df = pd.read_csv('./data/data_active.csv', index_col=0)
df
```

```python
df['track_id'].nunique()
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

n_neighbors = 10
radius = 1.0 # default
metric = 'cosine'
# more params excluded

knn = NearestNeighbors(n_neighbors=n_neighbors, radius=radius, metric=metric)
knn.fit(df.drop(labels=['track_id', 'track_name'], axis=1))
```

```python
track = '3GCdLUSnKSMJhs4Tj6CV3s'
coords = df.query('track_id == @track').drop(['track_id', 'track_name'], axis=1)
coords
```

```python
distance, index = knn.kneighbors(coords, 100)
index
```
